```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(data.drop('selling_price',axis=1),data['selling_price'],test_size=0.2,random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print('train mse: ',mean_squared_error(y_train,train_pred)**0.5)
print('test mse: ',mean_squared_error(y_test,test_pred)**0.5)
```

```python
from sklearn.model_selection import KFold
data.reset_index(drop=True,inplace=True)
data
```

```python
kf = KFold(n_splits = 5)
```

```python
X = data.drop('selling_price',axis=1)
y = data['selling_price']
```

```python
for i,j in kf.split(X):
  print(i,j)
```

```python
for train_index,test_index in kf.split(X):
  X_train,X_test = X.loc[train_index],X.loc[test_index]
  y_train,y_test = y[train_index],y[test_index] # 1줄이어서 loc 안사용
```

```python
train_rmse_total = []
test_rmse_total = []
```

```python
for train_index,test_index in kf.split(X):
  X_train,X_test = X.loc[train_index],X.loc[test_index]
  y_train,y_test = y[train_index],y[test_index]

  model = RandomForestRegressor()
  model.fit(X_train,y_train)
  train_pred = model.predict(X_train)
  test_pred = model.predict(X_test)

  train_mse = mean_squared_error(y_train,train_pred)**0.5
  test_mse = mean_squared_error(y_test,test_pred)**0.5

  train_rmse_total.append(train_mse)
  test_rmse_total.append(test_mse)
```

```python
print('train_rmse:', sum(train_rmse_total)/ 5)
```
