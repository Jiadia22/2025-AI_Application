```python
import pandas as pd
import glob # 경로 이름 저장해서 파일 처리 다룬다
import re # 정규식
from functools import reduce # 2차원을 1차원으로

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords #불용어 정해져 있음
from nltk.stem import WordNetLemmatizer # 단어형태일반화 - 표제어 추출

from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
all_files = glob.glob('/content/drive/MyDrive/Colab Notebooks/myCabinetExcelData*.xls')
all_files
```

```python
all_files_data = []

for file in all_files:
  data_frame = pd.read_excel(file)
  all_files_data.append(data_frame)

all_files_data[0]
```

```python
all_files_data_concat = pd.concat(all_files_data, axis=0, ignore_index=True)
all_files_data_concat
```

```python
all_files_data_concat.to_csv('/content/drive/MyDrive/Colab Notebooks/riss_bigdata.csv', encoding='utf-8', index=False)
```

```python
all_title = all_files_data_concat['제목']
all_title
```

```python
stopWords = set(stopwords.words("english")) #중복 없는 집합
lemma = WordNetLemmatizer()
```

```python
words = []

for title in all_title:
  EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title)) #알파벳으로 시작하지 않는 단어는 공백으로 바꿔라. 제거하라
  EnWordsToken = word_tokenize(EnWords.lower()) #소문자로 정규화
  EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords] #불용어 제거
  EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
  words.append(EnWordsTokenStopLemma) #표제어 추출
```

```python
print(words)
```

```python
words2 = list(reduce(lambda x, y: x + y, words)) #1차원으로 바뀜
print(words2)
```

```python
count = Counter(words2) #개수 추출
count
```

```python
word_count = dict()

for tag, counts in count.most_common(50): #상위 50개
  if(len(str(tag))>1): # 단어 길이가 1보다 큰거
    word_count[tag] = counts
    print("%s : %d" % (tag, counts))
```

```python
del word_count['big'] #너무 많아서 지움
del word_count['data']
```

```python
plt.figure(figsize=(12, 5))
plt.xlabel("word")
plt.ylabel("count")
plt.grid(True)

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True) #단어 개수가 많은거부터
sorted_Values = sorted(word_count.values(), reverse=True)

plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='85')

plt.show()
```

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color = 'ivory', stopwords=stopwords, width=800, height=600)
cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
