```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100) #-1 1 100 
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y_1)
ax[0, 1].plot(x, y_2)
ax[1, 0].plot(x, y_3)
ax[1, 1].plot(x, y_4)

plt.show()
```

```python
ax1 = plt.subplot(321)
plt.plot(x, y_1)
ax2 = plt.subplot(322)
plt.plot(x, y_1)
ax3 = plt.subplot(312)
plt.plot(x, y_1)
ax4 = plt.subplot(325)
plt.plot(x, y_1)
ax5 = plt.subplot(326)
plt.plot(x, y_1)

plt.show()
```

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color="#000000")
plt.plot(X_2, Y_2, c="c")

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b",linestyle='dashed')
plt.plot(X_2, Y_2, c="r",linestyle='dotted')
plt.title('Two Lines')
plt.show()
```

```python
plt.plot(X_1, Y_1, color='b', linestyle='dashed', label='line_1')
plt.plot(X_2, Y_2, color='r', linestyle='dotted', label='line_2')
plt.legend(shadow=True, fancybox=False, loc='upper right')

plt.title('$y = ax + b$')
plt.xlabel('$x_{lines}$')
plt.ylabel('$y_{line}$')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, ci=95) #95%
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

```python
sns.countplot(x='smoker', hue='time', data=tips)
```

```python
#y 
sns.barplot(x='day',y='total_bill',data=tips) #     .
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker', data=tips, palette='muted')
```

```python
sns.swarmplot(x='day',y='total_bill',hue='smoker',data=tips,palette='muted') #    .     .
```

```python
g = sns.FacetGrid(tips, col='time', row='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
g = sns.FacetGrid(tips, col='time', row='sex')
g.map(sns.histplot, 'total_bill')
```

plotly

```python
import plotly.express as px
df = px.data.iris()
df
```

```python
fig = px.scatter(df,x='sepal_width',y='sepal_length')
fig.show()
```

```python
fig = px.parallel_coordinates(
    df,
    color='species_id',
    labels={
        'species_id': 'Species',
        'sepal_width': 'Sepal Width',
        'sepal_length': 'Sepal Length',
        'petal_width': 'Petal Width',
        'petal_length': 'Petal Length',
    },
    color_continuous_scale=px.colors.diverging.Tealrose,
    color_continuous_midpoint=2
)

fig.show()
```
